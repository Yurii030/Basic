## distplot

```python
x= np.random.normal(size=100)
x
```

```python
sns.distplot(x, bins=20,kde=True,rug=True) #bins : 막대수
```

```python
sns.distplot(x, bins=20,kde=True,rug=False) #bins : 막대수
```

```python
sns.distplot(x, bins=20,kde=False,rug=True) #bins : 막대수
```

```python
sns.distplot(x, bins=20,kde=True,rug=True,hist_kws=dict(edgecolor='k',linewidth=0.5)) #bins : 막대수
```
