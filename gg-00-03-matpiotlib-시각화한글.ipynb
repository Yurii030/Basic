```python
#운영 체제마다 한글이 보이게 하는 설정 # 윈도우

import platform
from matplotlib import font_manager, rc 
import matplotlib

matplotlib.rcParams['axes.unicode_minus'] = False  #'-' 부호가 제대로 표시되게 하는 설정 

#운영 체제마다 한글이 보이게 하는 설정 # 윈도우

if platform.system() == 'Windows':  # -# 윈도우
    path = "c:\Windows\Fonts\malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name() 
    rc('font', family=font_name)
#맥
elif platform.system() == 'Darwin': 
    rc('font', family='AppleGothic')
#리눅스
elif platform.system() == 'Linux':
    rc('font', family='NanumBarunGothic')
```

```python
import pandas as pd
from pandas import DataFrame
```

```python
df = pd.read_csv('data/comma.txt', sep=',',encoding = 'cp949')
#df2 = DataFrame.from_csv('data/comma2.txt',sep=',')
```

```python
df
```

```python
graph = df.plot(kind ='bar', figsize=(20,5))  #figsize=(너비, 높이)

graph.set_title("기말성적표", fontsize=20)
graph.set_xlabel("이름")## x축 label
graph.set_ylabel("성적")## y축 label
graph.legend(loc='right') ;  # best, center, ## legend : 범례
```

```python
plt.style.use('ggplot') # ggplot style로 바꾸기

graph = df.plot(kind ='bar', figsize=(20,5))  #figsize=(너비, 높이)
graph.set_title("기말성적표", fontsize=20)
 
graph.set_xlabel("이름")# x축 label
graph.set_ylabel("성적")# y축 label
graph.legend(loc='right') ;  # best, center, ## legend : 범례
```
