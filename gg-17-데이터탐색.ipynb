### corr() - 상관관계 계수

- df.corr(method='pearson')
- method = pearson / kendall / spearman 의 적용할 상관계수 방식

```python
# 6x3짜리 객체를 생성

col1 = [1,2,3,4,5,6]
col2 = [1,4,2,8,16,32]
col3 = [6,5,4,3,2,1]
data = {"col1":col1,"col2":col2,"col3":col3}
df = pd.DataFrame(data)
print(df)
```

```python
print(df.corr(method='pearson'))

#          col1      col2      col3
# col1  1.000000  0.887739 -1.000000
# col2  0.887739  1.000000 -0.887739
# col3 -1.000000 -0.887739  1.000000

# col1이 증가할 경우 col2는 대체로 증가하는 경향을 가지기에 0<p<1의 값을 가지며, 양의 상관관계 계수
# col1이 증가할 경우 col3은 완벽히 감소하기 때문에 p=-1의 값을 가집, 음의 상관관계 계수
```
