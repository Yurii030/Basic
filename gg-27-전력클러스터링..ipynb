```python
power.head()
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc,font_manager
%matplotlib inline
from scipy.cluster.hierarchy import dendrogram , linkage
plt.figure(figsize=(10,5))
link_dist = linkage(power,metric='euclidean',method = 'ward')
dendrogram(link_dist,labels=power.index)
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4).fit(power)
print(kmeans.n_clusters)
centers =kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
power['클러스터']=kmeans.labels_
power
```

## DBSCAN : 밀도 기반 클러스터링

## two-moons 데이터

```python
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans, DBSCAN
X,y = make_moons(n_samples=300,noise=0.05,
random_state=11)
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
kmeans=KMeans(n_clusters=2)
predict = kmeans.fit_predict(X)
plt.scatter(X[:,0],X[:,1],c=predict)
```

```python
dbscan = DBSCAN(eps=0.2,min_samples=10,
metric='euclidean')
predict = dbscan.fit_predict(X)
plt.scatter(X[:,0],X[:,1],c=predict)
```

```python
dbscan = DBSCAN(eps=0.1,min_samples=5,
metric='euclidean')
predict = dbscan.fit_predict(X)
plt.scatter(X[:,0], X[:,1],c=predict)
```

```python
dbscan = DBSCAN(eps=0.1,min_samples=6,
metric='euclidean')
predict = dbscan.fit_predict(X)
plt.scatter(X[:,0], X[:,1],c=predict)
```

```python
dbscan = DBSCAN(eps=0.2,min_samples=5,
metric='euclidean')
predict = dbscan.fit_predict(X)
plt.scatter(X[:,0], X[:,1],c=predict)
```
