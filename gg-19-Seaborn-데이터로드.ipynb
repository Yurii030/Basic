## swarmplot

```python
sns.swarmplot(x="species",y="petal_width",data=iris)
```

```python
tips = sns.load_dataset("tips") #팁 데이터 세트
tips.head()
```

```python
sns.swarmplot(x="total_bill",y="tip",data=tips)
```

## stripplot

```python
# stripplot : data가 겹치면서 출력
# 카테고리 변수에 대한 구분
sns.stripplot(x="total_bill",y="sex",data=tips)
```

## barplot

```python
# barplot : 카테고리 실수의 값의 평균과 편차를 표시하는 바 차트, 막대의 높이: 평균
#ci =95
tips=sns.load_dataset("tips")
sns.barplot(x= "day",y="total_bill", data= tips);
```

```python
sns.barplot(x= "day",y="total_bill",hue="sex", data= tips);
plt.title("요일 별, 성별 전체 팁의 Violin plot")
plt.show()
```

```python
sns.boxplot(x= "day",y="total_bill",hue="sex", data= tips);
plt.title("요일 별, 성별 전체 팁의 Violin plot")
plt.show()
```

```python
sns.violinplot(x= "day",y="total_bill",hue="sex", data= tips);
plt.title("요일 별, 성별 전체 팁의 Violin plot")
plt.show()
```
