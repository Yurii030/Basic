## 시본기본설정(seaborn: Aesthetics)

## 환경설정 파라미터 설정에는 두가지 종류
- Styling
- Scaling

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set() # 디폴트로 사용
%matplotlib inline
```

```python
tips = sns.load_dataset("tips") #팁 데이터 세트
tips
```

## boxplot(Box and Whisker Plot) : 상자 수염 그림
### 0~100 :101개 자료

```python
sns.set_style("whitegrid") # set_style : darkgrid,whitegrid,dark,white,ticks에 사용되는

# boxplot  : 박스- 휘스키 플롯 -> 박스와 박스 바깥의 선(whisker)으로 이루어진~
sns.boxplot(x="day",y="total_bill",data=tips)

plt.title("day,total_tip") # boxplot : 한글지원 에러
plt.show()
```

```python
sns.set_style("ticks") # ticks -> 배경이 사라진다

# boxplot  : 박스- 휘스키 플롯 -> 박스와 박스 바깥의 선(whisker)으로 이루어진~
sns.boxplot(x="day",y="total_bill",data=tips)

plt.title("day,total_tip") # boxplot : 한글지원 에러
plt.show()
```

```python
sns.set_style("dark") 

# boxplot  : 박스- 휘스키 플롯 -> 박스와 박스 바깥의 선(whisker)으로 이루어진~
sns.boxplot(x="day",y="total_bill",data=tips)

plt.title("day,total_tip") # boxplot : 한글지원 에러
plt.show()
```

```python
sns.set_style("ticks") 
sns.boxplot(x="day",y="total_bill",data=tips)
sns.despine()
```

```python
sns.set_style("ticks") 
sns.boxplot(x="day",y="total_bill",data=tips)
sns.despine(left=True) # 기본 위 , 오른 쪽 삭제 + left = True
```

## lmplot

```python
# column간의 선형관계 확인, Implot()hue옵션 사용 가능, legplot() -> hue 옵션
sns.lmplot(x="total_bill",y="tip",height=8,data=tips)
plt.show()
```

```python
# column간의 선형관계 확인, Implot()hue옵션 사용 가능, legplot() -> hue 옵션
sns.lmplot(x="total_bill",y="tip",height=8,ci=0,data=tips) # 신뢰도 옵션 ci
plt.show()
```

```python
# column간의 선형관계 확인, Implot()hue옵션 사용 가능, legplot() -> hue 옵션
sns.lmplot(x="total_bill",y="tip",height=8,ci=100,data=tips) # 신뢰도 옵션 ci
plt.show()
```

```python
# column간의 선형관계 확인, Implot()hue옵션 사용 가능, legplot() -> hue 옵션
sns.lmplot(x="total_bill",y="tip",height=8,ci=50,data=tips) # 신뢰도 옵션 ci
plt.show()
```

```python
# column간의 선형관계 확인, Implot()hue옵션 사용 가능, legplot() -> hue 옵션
sns.lmplot(x="total_bill",y="tip",hue="smoker",height=8,data=tips) 
plt.show()
```

```python
sns.lmplot(x="total_bill",y="tip",hue="smoker",col='day',col_wrap=2,height=6,data=tips) 
plt.show()
```

## Jointplot()

```python
#산점도와 분포를 동시에 그려주며 데이터만 표현 가능.
sns.jointplot(x="total_bill",y="tip",height=8,data=tips)
plt.show()
```
