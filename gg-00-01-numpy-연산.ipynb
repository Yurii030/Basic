# np.random.rand()

```python
# 0과 1사이의 값에서 난수 어레이를 생성
rand_a = np.random.rand()
rand_b = np.random.rand(5)
rand_c = np.random.rand(2,3)

print(rand_a)
print("*"*70)

print(rand_b)
print("*"*70)

print(rand_c)
```

# np.random.randn()


```python
#-1에서 1 사이 표준 정규분포(standard normal disrtibution)으로 부터 난수 값을 반환

randn_a = np.random.randn()
print(randn_a)
print("*"*70)

randn_b = np.random.randn(5) #5개
print(randn_b)
print("*"*70)

randn_c = np.random.randn(100) #100개
print(randn_c)
print("*"*70)

randn_d = np.random.randn(2,3) #2행 3열
print(randn_d)
```

```python
# (a,b,size) a~b-1 범위에서 임의의 "정수"를 만듬, a를 생략하면 0부터 시작
randint_a = np.random randint(2,size=5)  # (최소값,최대값,size) -> 최소값~최대값)
print(randint_a)
print("*"*70)

randint_b = np.random randint(2,3,size=7)  #size 2를 7개 
print(randint_b)
print("*"*70)
```
