## 로지스틱회귀분석(Logistic regression)

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
%matplotlib inline
```

```python
# 합격한 공부시간
pass_time = [3, 8, 9, 9,  9.5, 10, 12, 14, 14.5, 15, 16, 16, 16.5, 17, 17, 17, 17.5,20, 20, 20] # 20개~

# 불합격한 공부시간
fail_time = [1, 2, 2.1,2.6, 2.7,2.8,2.9, 3, 3.2,3.4,3.5,3.6, 3, 5, 5.2, 5.4,6,6.5,7,8] # 20개~
```

```python
from sklearn.linear_model import LogisticRegression
X=np.hstack((pass_time,fail_time))
X
```

```python
y1=[1]*len(pass_time)
y0=[0]*len(fail_time)
y=np.hstack((y1,y0))
(X,y)
```

```python
import matplotlib.pyplot as plt
fig=plt.figure(figsize(8,6))
plt.xlim(0,21)
plt.ylim(-0.1,1.1)
plt.xlabel("Study time")
plt.ylabel("Pass time")
plt.scatter(X,y)
plt.show()
```

```python
def logreg(z):
    return 1 / (1+np.exp(-z))
fig=plt.figure(figsize=(6,4))
plt.xlim(-10,10)
plt.ylim(-0.1,1.1)
a_list=[0.3,0.5,1]
b_list=[0]
XX=np.linspace(-10,10,100)
for a in a_list :
    for b in b_list :
        yy=logreg(a*XX+b)
        plt.plot(XX,yy,c='k')
```

```python
def logreg(z):
    return 1 / (1+np.exp(-z))
fig=plt.figure(figsize=(6,4))
plt.xlim(-10,10)
plt.ylim(-0.1,1.1)
a_list=[0.8]
b_list=[-2,0,2]
XX=np.linspace(-10,10,100)
for a in a_list :
    for b in b_list :
        yy=logreg(a*XX+b)
        plt.plot(XX,yy,c='k')
```

```python
model= LogisticRegression()
model.fit(X.reshape(-1,1),y.reshape(-1,1))
print(model.coef_)
print(model.intercept_)
```
